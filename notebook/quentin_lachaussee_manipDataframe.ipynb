```python
import pandas as pd

chemin_Buchanan = "./../data/dictionary/1757_Buchanan-J.csv"
df_Buchanan = pd.read_csv(chemin_Buchanan, sep=";")
df_Buchanan.head(10)
```

```python
chemin_PronChar = "./../data/dictionary/Buchanan_PronChar_counts.csv" 
df_PronChar = pd.read_csv(chemin_PronChar, sep=";")
print("Max nb value", df_PronChar['char'].value_counts().max())
df_PronChar["graph"] = df_PronChar["graph"].fillna("")
df_PronChar
```

```python
def split_pron(pron, df_PronChar):
    df_PronChar
    
    message = ""
    for value in pron:
        res = df_PronChar.graph[df_PronChar['char']==value].values[0]
        message+=str(res)
    return message
```

```python
df_Buchanan['pron'].head(10).apply(lambda x: split_pron(x, df_PronChar))
```

```python
df_Buchanan['motVedette '] = df_Buchanan['pron'].apply(lambda x: split_pron(x, df_PronChar))
```
