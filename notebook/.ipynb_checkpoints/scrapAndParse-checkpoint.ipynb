```python
finded = soupletteAside.findAll('div',{"class":"property"})

for v in finded:
    #print(v)
    #print(">>>>>",v.find('h4').text)
    vH4 = v.find('h4').text.strip()
    print(vH4)
```

```python
for v in finded:
    vH4 = v.find('h4').text.strip()
    if "Identifiant" in vH4:
        print(vH4)
        
        soupletteV = v.find('div',{"class":"value"})
        print(soupletteV)
        print(soupletteV.text)
        
        vJPG = soupletteV.text.strip()
        print(vJPG)
```

```python
import requests
urlTarget = "https://www.linguee.fr/anglais-francais/traduction/love.html"
response = requests.get(urlTarget)
from bs4 import BeautifulSoup
love = BeautifulSoup(response.text, 'html.parser')
love_boddy = love.find('body')
love_boddy
```

```python
from IPython import display

dict_audio = {i: f"https://www.linguee.fr/mp3/{link['id']}.mp3" for i, link in enumerate(love_boddy.findAll('a',{'class':'audio'}))}
for link in dict_audio.values():
    d = display.Audio(url=link)
d
```

```python
love_aside = love_boddy.findAll('span',{"class":"tag_lemma"})
love_aside
```

```python
import pandas as pd
dict_audio = {}
df = pd.DataFrame(columns=["Name", "Link", "Tag"])
for i,v in enumerate(love_aside):
    name = v.findAll('a',{"class":"dictLink"})
    link = v.findAll('a',{"class":"audio"})
    tag = v.findAll('span',{"class":"tag_wordtype"})
    if len(name) + len(link) == 2:
        dict_audio[name[0].text] = link[0]['id']
    if len(name) + len(link) + len(tag) == 3:
        df.loc[i] = [name[0].text, link[0]['id'], tag[0].text]
```

```python
dict_audio
```

```python
df
```

```python
import ipywidgets as widgets
list_widget = []
for l in df.Link:
    audio_url = f"https://www.linguee.fr/mp3/{l}.mp3"
    list_widget.append(widgets.Audio(url=audio_url))

widgets.VBox(list_widget)
```
