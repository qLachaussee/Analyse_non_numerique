## 0. Import des librairies

```python
import pandas as pd
```

## 1. Mot vedette

```python
chemin_Buchanan = "./../data/dictionary/1757_Buchanan-J.csv"
df_Buchanan = pd.read_csv(chemin_Buchanan, sep=";")
df_Buchanan.head(10)
```

```python
chemin_pron_char = "./../data/dictionary/Buchanan_PronChar_counts.csv" 
df_pron_char = pd.read_csv(chemin_pron_char, sep=";")
df_pron_char["graph"] = df_pron_char["graph"].fillna("")
df_pron_char
```

```python
def split_pron(pron, df):    
    message = ""
    for value in pron:
        res = df.graph[df['char']==value].values[0]
        message+=str(res)
    return message
```

```python
df_Buchanan['pron'].head(10).apply(lambda x: split_pron(x, df_pron_char))
```

```python
%%time
df_Buchanan['motVedette'] = df_Buchanan['pron'].apply(lambda x: split_pron(x, df_pron_char))
```

```python
df_Buchanan
```

## 2. Pron Sampa

```python
chemin_Bigi = "./../data/dictionary/2015_Bigi-B.dict" 
df_Bigi = pd.read_csv(chemin_Bigi, delimiter=" \[\] ", encoding="utf8", engine='python', names=['hwd','pronSampa'], skiprows=8 )
df_Bigi.head(5)
```

```python
df_final = df_Buchanan.merge(df_Bigi, left_on="motVedette", right_on="hwd")
```

```python
df_final
```

## 3. Pron IPA

```python
chemin_sampa_ipa = "./../data/dictionary/sampa-ipa.txt" 
df_sampa_ipa = pd.read_csv(chemin_sampa_ipa, delimiter=" +", encoding="utf8", engine='python', names=['sampa','ipa'])
df_sampa_ipa.drop_duplicates(subset=['sampa'], keep='first', inplace=True)
df_sampa_ipa.head(5)
```

```python
def sampa2ipa(sampa, df):
    message = ""
    list_val_sampa = sampa.split(" ")
    for val in list_val_sampa:
        df_line = df[df['sampa']==val]
        if len(df_line==1):
            message += df_line.iloc[0]['ipa']
        else:
            message += '?'
    return message
```

```python
df_final.head(10).apply(lambda x: sampa2ipa(x['pronSampa'], df_sampa_ipa),1)
```

```python
%%time
df_final['pronIpa'] = df_final.apply(lambda x: sampa2ipa(x['pronSampa'], df_sampa_ipa),1)
```

```python
df_final
```

## 4. Pron POS

```python
def switch_pos(argument):
    switcher = {
        "n.": "noun",
        "a.": "adjective",
        "v.": "verb",
        "p.": "pronoun"
    }
    return switcher.get(argument, "?")
```

```python
df_final["POS"].head(10).apply(lambda x: switch_pos(x))
```

```python
%%time
df_final["pronPos"] = df_final["POS"].apply(lambda x: switch_pos(x))
```
